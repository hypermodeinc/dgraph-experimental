```python
import networkx as nx
import matplotlib.pyplot as plt
# Generate the Barabási-Albert graph

import random

# Create a directed graph
G = nx.DiGraph()

# Add nodes
num_nodes = 10
G.add_nodes_from(range(num_nodes))

# Add edges with varying out-degrees
for node in range(num_nodes):
    # Randomly choose the number of outgoing edges for each node (1 to num_nodes-1)
    num_out_edges = random.randint(1, num_nodes - 1)
    # Randomly select target nodes, avoiding self-loops
    targets = random.sample([n for n in range(num_nodes) if n != node], num_out_edges)
    for target in targets:
        G.add_edge(node, target)

# Draw the graph
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=500, node_color="skyblue", edge_color="gray", font_size=10)
```

```python
with open("graph.edgelist", "rb") as f:
    G = nx.read_edgelist("graph.edgelist", nodetype=int)
```

```python
# Load the graph from the pickle file

   
pagerank_scores = nx.pagerank(G, alpha=0.85, max_iter=100, tol=1e-8)

# Display the PageRank scores
print("Node : PageRank Score")
for node, score in pagerank_scores.items():
    print(f"{node} : {score:.6f}")
# Scale node sizes based on PageRank scores
node_sizes = [score * 5000 for score in pagerank_scores.values()]  # Scale factor to make sizes visually distinguishable

# Draw the graph
plt.figure(figsize=(12, 10))
nx.draw(
    G,
    with_labels=True,
    node_size=node_sizes,
    node_color="skyblue",
    edge_color="gray",
    font_size=10,
    font_color="darkblue"
)
plt.title("Barabási-Albert Graph with Node Sizes Based on PageRank", fontsize=16)
plt.show()

# Verify
```
